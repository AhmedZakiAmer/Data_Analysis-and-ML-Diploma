## Task 1&2&3 #Read the file

```python
import numpy as np
import pandas as pd
filename = 'D:\data.csv'
data = pd.read_csv(filename)
data
```

```python
#Task 4->Change column names to capitalize
data.columns = [i.upper() for i in data.columns]
data.columns
```

```python
#Task 5->Change column (social type) to (MARITAL_STATUS)
data.rename(columns={'SOCIALTYPE':'MARITAL_STATUS'},inplace = True)
data
```

## Task 10 deal with nulls

```python
data['EDUCATION'].fillna(method='ffill',inplace = True)
data['CITY'].fillna(method='ffill',inplace = True)
data['SALARY'].fillna(data['SALARY'].mean(),inplace = True)
data.info()
```

## Task 11 Treat outliers 

```python
data['SALARY'] = np.where(data['SALARY'] > (data['SALARY'].quantile([0.75]).values[0] + data['SALARY'].mean()) * 1.5,
                         (data['SALARY'].quantile([0.75]).values[0] + data['SALARY'].mean()) * 1.5,
                         data['SALARY'])
data['SALARY'].describe()
```

```python
#Task 12 How many average salaries by year
y = data.pivot_table(index='YEAR'
               ,values=['SALARY']
               ,aggfunc=np.mean
          )
y
```

```python
#Task 13 How many average salaries by city
x= data.pivot_table(index='CITY'
               ,values=['SALARY']
               ,aggfunc=np.mean
          )
# x.reset_index()
x
```

```python
#Task 14 most city has salaries 
x['SALARY'].idxmax()
```

```python
#Task 15 most education
data['EDUCATION'].value_counts().idxmax()
```
